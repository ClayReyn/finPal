```python
def daybyday_solo(dbdf, testfilters_df, testfilter_irow, daterange):
    techcol = testfilters_df.at[testfilter_irow, 'techcol']
    techleft = testfilters_df.at[testfilter_irow, 'techleft']
    techright= testfilters_df.at[testfilter_irow, 'techright']

    daybyday_rows = []

    testname = f'{techcol}'

    for adate in daterange:
        obsv = len( 
            dbdf.loc[ 
                ( 
                    dbdf[techcol].between(
                        left= techleft,
                        right= techright
                    )
                )
            ]
        )
        Ov0 = len( 
            dbdf.loc[ 
                ( 
                    dbdf[techcol].between(
                        left= techleft,
                        right= techright
                    ) 
                ) & ( 
                    dbdf['close_chngpct_24_tmro'] > 0
                )
            ]
        )
        Ov05 = len( 
            dbdf.loc[ 
                ( 
                    dbdf[techcol].between(
                        left= techleft,
                        right= techright
                    ) 
                ) & ( 
                    dbdf['close_chngpct_24_tmro'] > 0.5
                )
            ]
        )
        Ov1 = len( 
            dbdf.loc[ 
                ( 
                    dbdf[techcol].between(
                        left= techleft,
                        right= techright
                    ) 
                ) & ( 
                    dbdf['close_chngpct_24_tmro'] > 1
                )
            ]
        )
        obsv_tickers = len( 
            dbdf.loc[ 
                ( 
                    dbdf[techcol].between(
                        left= techleft,
                        right= techright
                    ) 
                ) 
            ]['ticker'].tolist()
        )
        Ov0_tickers = len( 
            dbdf.loc[ 
                ( 
                    dbdf[techcol].between(
                        left= techleft,
                        right= techright
                    ) 
                ) & ( 
                    dbdf['close_chngpct_24_tmro'] > 0
                )
            ]['ticker'].tolist()
        )
        daybyday_rows.append([testname, adate, obsv, Ov0, Ov05, Ov1, obsv_tickers, Ov0_tickers])
    daybyday_df = pd.DataFrame(data=daybyday_rows, columns=['test', 'date', 'obsv', 'Ov0', 'Ov05', 'Ov1', 'obsv_tickers', 'Ov0_tickers'])
    return daybyday_df
```

```python
def daybyday_2(dbdf, testfilters_df, testfilter_irow, daterange):
    techcol = testfilters_df.at[testfilter_irow, 'techcol']
    techleft = testfilters_df.at[testfilter_irow, 'techleft']
    techright= testfilters_df.at[testfilter_irow, 'techright']
    tech2 = testfilters_df.at[testfilter_irow, 'tech2']
    tech2left = testfilters_df.at[testfilter_irow, 'tech2left']
    tech2right= testfilters_df.at[testfilter_irow, 'tech2right']

    daybyday_rows = []

    testname = f'{techcol}_+_{tech2}'

    for adate in daterange:
        obsv = len( 
            dbdf.loc[ 
                ( 
                    dbdf[techcol].between(
                        left= techleft,
                        right= techright
                    )
                ) & ( 
                  dbdf[tech2].between( 
                      left= tech2left,
                      right= tech2right
                    )
                ) & ( 
                    dbdf['date'] == adate
                )
            ]
        )
        Ov0 = len( 
            dbdf.loc[ 
                ( 
                    dbdf[techcol].between(
                        left= techleft,
                        right= techright
                    ) 
                )  & ( 
                    dbdf[tech2].between( 
                        left= tech2left,
                        right= tech2right
                    )
                ) & ( 
                    dbdf['close_chngpct_24_tmro'] > 0
                ) & ( 
                    dbdf['date'] == adate
                )
            ]
        )
        Ov05 = len( 
            dbdf.loc[ 
                ( 
                    dbdf[techcol].between(
                        left= techleft,
                        right= techright
                    ) 
                )  & ( 
                    dbdf[tech2].between( 
                        left= tech2left,
                        right= tech2right
                    )
                ) & ( 
                    dbdf['close_chngpct_24_tmro'] > 0.5
                ) & ( 
                    dbdf['date'] == adate
                )
            ]
        )
        Ov1 = len( 
            dbdf.loc[ 
                ( 
                    dbdf[techcol].between(
                        left= techleft,
                        right= techright
                    ) 
                ) & ( 
                    dbdf[tech2].between( 
                        left= tech2left,
                        right= tech2right
                    )
                ) & ( 
                    dbdf['close_chngpct_24_tmro'] > 1
                ) & ( 
                    dbdf['date'] == adate
                )
            ]
        )
        obsv_tickers = ( 
            dbdf.loc[ 
                ( 
                    dbdf[techcol].between(
                        left= techleft,
                        right= techright
                    ) 
               ) & ( 
                   dbdf[tech2].between( 
                       left= tech2left,
                       right= tech2right
                   )
               ) & ( 
                    dbdf['date'] == adate
                )
            ]['ticker'].tolist()
        )
        Ov0_tickers = ( 
            dbdf.loc[ 
                ( 
                    dbdf[techcol].between(
                        left= techleft,
                        right= techright
                    ) 
                ) & ( 
                    dbdf[tech2].between( 
                        left= tech2left,
                        right= tech2right
                    )
                ) & ( 
                    dbdf['close_chngpct_24_tmro'] > 0
                ) & ( 
                    dbdf['date'] == adate
                )
            ]['ticker'].tolist()
        )
        if obsv > 0:
            Ov0c = np.around(Ov0 / obsv, 4)
            Ov05c = np.around(Ov05 / obsv, 4)
            Ov1c = np.around(Ov1 / obsv, 4)
        else: 
            Ov0c = np.NaN
            Ov05c = np.NaN
            Ov1c = np.NaN
        daybyday_rows.append([testname, adate, obsv, Ov0, Ov0c, Ov05, Ov05c, Ov1, Ov1c, obsv_tickers, Ov0_tickers])
    daybyday_df = pd.DataFrame(data=daybyday_rows, columns=['test', 'date', 'obsv', 'Ov0', 'Ov0c', 'Ov05', 'Ov05c', 'Ov1', 'Ov1c', 'obsv_tickers', 'Ov0_tickers'])
    return daybyday_df
    
```

```python
def daybyday_4(dbdf, testfilters_df, testfilter_irow, daterange):
    pick1 = testfilters_df.at[testfilter_irow, 'pick1']
    pick1left = testfilters_df.at[testfilter_irow, 'pick1left']
    pick1right = testfilters_df.at[testfilter_irow, 'pick1right']
    pick2 = testfilters_df.at[testfilter_irow, 'pick2']
    pick2left = testfilters_df.at[testfilter_irow, 'pick2left']
    pick2right = testfilters_df.at[testfilter_irow, 'pick2right']
    techcol = testfilters_df.at[testfilter_irow, 'techcol']
    techleft = testfilters_df.at[testfilter_irow, 'techleft']
    techright = testfilters_df.at[testfilter_irow, 'techright']
    tech2 = testfilters_df.at[testfilter_irow, 'tech2']
    tech2left = testfilters_df.at[testfilter_irow, 'tech2left']
    tech2right = testfilters_df.at[testfilter_irow, 'tech2right']
    
    daybyday_rows = []

    for adate in daterange: 
        obsv = len( 
            dbdf.loc[ 
                    (
                        dbdf[pick1].between(
                            left= pick1left,
                            right= pick1right
                        )
                    ) & ( 
                        dbdf[pick2].between( 
                            left= pick2left,
                            right= pick2right
                        )
                    ) & ( 
                        dbdf[techcol].between( 
                            left= techleft,
                            right= techright
                        )
                    ) & ( 
                        dbdf[tech2].between( 
                            left= tech2left,
                            right= tech2right
                        )
                    ) & ( 
                        dbdf['date'] == adate
                    )
            ]
        )
        Ov0 = len( 
            dbdf.loc[ 
                    (
                        dbdf[pick1].between(
                            left= pick1left,
                            right= pick1right
                        )
                    ) & ( 
                        dbdf[pick2].between( 
                            left= pick2left,
                            right= pick2right
                        )
                    ) & ( 
                        dbdf[techcol].between( 
                            left= techleft,
                            right= techright
                        )
                    ) & ( 
                        dbdf[tech2].between( 
                            left= tech2left,
                            right= tech2right
                        )
                    ) & ( 
                        dbdf['date'] == adate
                    ) & ( 
                        dbdf['close_chngpct_24_tmro'] > 0
                    )
            ]
        )
        obsv_tickers = ( 
            dbdf.loc[ 
                    (
                        dbdf[pick1].between(
                            left= pick1left,
                            right= pick1right
                        )
                    ) & ( 
                        dbdf[pick2].between( 
                            left= pick2left,
                            right= pick2right
                        )
                    ) & ( 
                        dbdf[techcol].between( 
                            left= techleft,
                            right= techright
                        )
                    ) & ( 
                        dbdf[tech2].between( 
                            left= tech2left,
                            right= tech2right
                        )
                    ) & ( 
                        dbdf['date'] == adate
                    )
            ]['ticker'].tolist()
        )
        Ov0_tickers = (
            dbdf.loc[ 
                    (
                        dbdf[pick1].between(
                            left= pick1left,
                            right= pick1right
                        )
                    ) & ( 
                        dbdf[pick2].between( 
                            left= pick2left,
                            right= pick2right
                        )
                    ) & ( 
                        dbdf[techcol].between( 
                            left= techleft,
                            right= techright
                        )
                    ) & ( 
                        dbdf[tech2].between( 
                            left= tech2left,
                            right= tech2right
                        )
                    ) & ( 
                        dbdf['date'] == adate
                    ) & ( 
                        dbdf['close_chngpct_24_tmro'] > 0
                    )
            ]['ticker'].tolist()
        )
        daybyday_rows.append(['test', adate, obsv, Ov0, obsv_tickers, Ov0_tickers])
    daybyday_df = pd.DataFrame(data=daybyday_rows, columns=['test', 'date', 'obsv', 'Ov0', 'obsv_tickers', 'Ov0_tickers'])
    return daybyday_df
```

```python
def _strfdateArray(stISOf, endISOf) -> list:
        ''' stISOf and endISOf take "YYYY-MM-DD" as strings. '''
        fromDate = date.fromisoformat(stISOf)
        toDate = date.fromisoformat(endISOf)
        aDay = timedelta(days=1)
        prevDate = toDate - aDay
        date_array = [ datetime.strftime(prevDate, '%Y-%m-%d') ]
        while prevDate > fromDate:
            prevDate = prevDate - aDay
            date_array.append(datetime.strftime(prevDate, '%Y-%m-%d'))
        return date_array
```
